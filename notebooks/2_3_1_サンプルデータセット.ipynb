### データセットの内容
- 特徴量（8つの特徴量）：
  - `MedInc`: 地区の世帯の中央値収入
  - `HouseAge`: 地区の住宅の中央値の築年数
  - `AveRooms`: 1世帯あたりの平均部屋数
  - `AveBedrms`: 1世帯あたりの平均寝室数
  - `Population`: 地区の人口
  - `AveOccup`: 1世帯あたりの平均居住者数
  - `Latitude`: 地区の緯度
  - `Longitude`: 地区の経度
- 目標変数：
  - `MedHouseVal`: 地区の住宅価格の中央値
